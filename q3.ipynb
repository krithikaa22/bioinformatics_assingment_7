```python
import pandas as pd

amino_values = {
    'Ala': [8.47, 8.95],
    'Cys': [1.39, 0.47],
    'Asp': [5.97, 5.91],
    'Glu': [6.32, 4.78],
    'Thr': [5.79, 6.54],
    'Phe': [3.91, 3.68],
    'Gly': [7.82, 8.54],
    'His': [2.26, 1.25],
    'Ile': [5.71, 4.77],
    'Lys': [5.76, 4.93],
    'Leu': [8.48, 8.78],
    'Met': [2.21, 1.56],
    'Asn': [4.54, 5.74],
    'Trp': [1.44, 1.24],
    'Pro': [4.63, 3.74],
    'Gln': [3.82, 4.75],
    'Arg': [4.93, 5.24],
    'Ser': [5.94, 8.05],
    'Tyr': [3.58, 4.13],
    'Val': [7.02, 6.76]
}


amino_acids_dict = {
    'Ala': 'A',
    'Arg': 'R',
    'Asn': 'N',
    'Asp': 'D',
    'Cys': 'C',
    'Gln': 'Q',
    'Glu': 'E',
    'Gly': 'G',
    'His': 'H',
    'Ile': 'I',
    'Leu': 'L',
    'Lys': 'K',
    'Met': 'M',
    'Phe': 'F',
    'Pro': 'P',
    'Ser': 'S',
    'Thr': 'T',
    'Trp': 'W',
    'Tyr': 'Y',
    'Val': 'V'
}


input_sequences = [
    'RATPTRWPVGCFNRPWTKWSYDEALDGIKAAGYAWTGLLTASKPSLHHATATPEYLAALKQKSRHAA',
    'AAAVMMGLAAIGAAIGIGILGGKFLEGAARQPDLIPLLRTQFFIVMGLVDAIPMIAVGLGLYVMFAVA',
    'AADVSAAVGATGQSGMTYRLGLSWDWDKSWWQTSTGRLTGYWDAGYTYWEGGDEGAGKHSLSFAPVFVYEFAGDSIKPFIEAGIGVAAFSGTRVGDQNLGSSLNFEDRIGAGLKFANGQSVGVRAIHYSNAGLKQPNDGIESYSLFYKIPI'
]

table = pd.DataFrame(columns=['A', 'B', 'Group'])
composition_table = pd.DataFrame(columns=amino_acids_dict.values())

for i in input_sequences:
    aas = list(i)
    composition_table = composition_table.append(pd.Series({aa: aas.count(aa)*100/len(aas) for aa in amino_acids_dict.values()}, name=input_sequences.index(i)))
display(composition_table)
print(amino_acids_dict.keys())
for i in input_sequences:
    a_val = sum([abs(composition_table.loc[input_sequences.index(i)][amino_acids_dict[aa]] - amino_values[aa][0]) for aa in amino_acids_dict.keys()])
    b_val = sum([abs(composition_table.loc[input_sequences.index(i)][amino_acids_dict[aa]] - amino_values[aa][1]) for aa in amino_acids_dict.keys()])
    if a_val < b_val:
        table = table.append(pd.Series({'A': a_val, 'B': b_val, 'Group': 'A' }, name=input_sequences.index(i)))
    else:
        table = table.append(pd.Series({'A': a_val, 'B': b_val, 'Group': 'B' }, name=input_sequences.index(i)))

table
```
